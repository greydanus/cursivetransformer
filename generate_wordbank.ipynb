```python
import numpy as np
import string

def generate_word_bank(num_examples=5000):
    letters = string.ascii_lowercase
    letter_probs = np.array([8.2, 1.5, 2.8, 4.3, 13, 2.2, 2, 6.1, 7, 0.15, 0.77, 4, 2.4, 6.7, 7.5, 1.9, 0.095, 6, 6.3, 9.1, 2.8, 0.98, 2.4, 0.15, 2, 0.074])
    letter_probs = np.maximum(letter_probs, 1)
    letter_probs /= letter_probs.sum()

    word_bank = []
    for _ in range(num_examples):
        length = np.random.randint(3, 12)
        word = ''.join(np.random.choice(list(letters), size=length, p=letter_probs))
        word = word.upper() if np.random.random() < 0.15 else (word.capitalize() if np.random.random() < 0.5 else word)
        word += '.' if np.random.random() < 0.1 else ''
        word_bank.append(word)

    return word_bank

def write_word_bank_to_file(word_bank, filename='word_bank.txt'):
    with open(filename, 'w') as f:
        f.write('const words = [' + ', '.join(f'"{w}"' for w in word_bank) + '];')

word_bank = generate_word_bank()
write_word_bank_to_file(word_bank)
```
